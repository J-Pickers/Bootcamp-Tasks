```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
raw_data = pd.read_csv("insurance.csv")
raw_data.head() # Test to see if import complete
```

```python
# Assigning Age column to x, and Charges column to y.
# Used .values to only pull the data and not the headers too
x = raw_data.iloc[:,:1].values
y = raw_data.iloc[:,-1].values

# Plotting a scatter plot with previous data, with red data points.
plt.scatter(x,y,color = "r")
# Labelling each axis
plt.xlabel("Age")
plt.ylabel("Charges")

plt.show()
```

```python
# Fitting linear regression model to data
l_reg_model = LinearRegression()
l_reg_model.fit(x,y)
```

```python
# Generating data for line of best fit, based on input values of x
best_fit = l_reg_model.predict(x)

# Plotting scatter plot once again, with best-fit line as well in blue
plt.scatter(x,y,color = "r")
plt.xlabel("Age")
plt.ylabel("Charges")
plt.plot(x,best_fit,color="b")

plt.show()
```

```python
# Choosing a value to predict for
prediction = [[66]]

# Appending this value to our data set, using .reshape to ensure array remains 2D
x_predict = np.append(x,prediction).reshape(-1,1)
# Generating predictions for this new set that contains the new value
y_predict = l_reg_model.predict(x_predict)

# Plotting once again
plt.scatter(x,y,color = "r")
plt.xlabel("Age")
plt.ylabel("Charges")
plt.plot(x_predict,y_predict,color="b")

plt.show()

# print output prediction specifically for the appended data point
print("Charges for a 66 year old expected to be: ",l_reg_model.predict(prediction))
```

```python
# Extrapolating the prediction further
prediction = [[75]]

# Appending this value to our data set, using .reshape to ensure array remains 2D
x_predict = np.append(x,prediction).reshape(-1,1)
# Generating predictions for this new set that contains the new value
y_predict = l_reg_model.predict(x_predict)

# Plotting once again
plt.scatter(x,y,color = "r")
plt.xlabel("Age")
plt.ylabel("Charges")
plt.plot(x_predict,y_predict,color="b")

plt.show()

# print output prediction specifically for the appended data point
print("Charges for a 75 year old expected to be: ",l_reg_model.predict(prediction))
```
