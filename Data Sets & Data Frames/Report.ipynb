```python
# Import pandas
import pandas as pd
```

```python
# Create a DataFrame with balance.txt
df = pd.read_csv("balance.txt",sep=" ")
```

Write the code needed to produce a report that provides the following information:
* Compare the average income based on ethnicity.
* On average, do married or single people have a higher balance?
* What is the highest income in our dataset?
* What is the lowest income in our dataset?
* How many cards do we have recorded in our dataset? (Hint: use `sum()`)
* How many females do we have information for vs how many males? (Hint: use `count()` for a list of all methods for computation of descriptive stats, explore the [pandas documentation](https://pandas.pydata.org/pandas-docs/stable/reference/frame.html#computations-descriptive-stats)).




```python
# Compare the average income based on Ethnicity

# Assigning variables to 3 refined data frames, filtered by Ethnicity
af_am = df[df["Ethnicity"] == "African American"]
asian = df[df["Ethnicity"] == "Asian"]
cauca = df[df["Ethnicity"] == "Caucasian"]

# Finding the mean income of these 3 frames, 
# Rounded to 3 decimal places to be consistent with source data.
avg_af_am = round(af_am["Income"].mean(), 3)
avg_asian = round(asian["Income"].mean(), 3)
avg_cauca = round(cauca["Income"].mean(), 3)

# f-strings to print out the results
print(f"The average income for African Americans is {avg_af_am}")
print(f"The average income for Asians is {avg_asian}")
print(f"The average income for Caucasians is {avg_cauca}")
```

```python
# Alternatively, if the formatting of the output is not as important,
# the following code can be used to retrieve the data more efficiently;

print(df.groupby("Ethnicity")["Income"].mean())
# This uses .groupby() to retrieve the mean income for all Ethnicity 
# groups in one line
```

```python
# On average, do married or single people have a higher balance?

# Creating 2 refined frames based on marriage status
married = df[df["Married"] == "Yes"]
single = df[df["Married"] == "No"]

# Finding the mean balance for both frames,
# rounded to 6 decimal places to be more readable
avg_married = round(married["Balance"].mean(), 6)
avg_single = round(single["Balance"].mean(), 6)

# f-strings to print the results
print(f"The average balance for a married person is {avg_married}")
print(f"The average balance for a single person is {avg_single}")
print("On average, single people have a higher balance.")
```

```python
# Again, for efficiency over presentation, this code can be used;

print(df.groupby("Married")["Balance"].mean())
```

```python
# What is the highest income in our dataset?

# Simply using the .max() function on the income column
max = df["Income"].max()
print(f"{max} is the highest income in the data.")

# What is the lowest income in our dataset?

# Same as before, but .min() instead
min = df["Income"].min()
print(f"{min} is the lowest income in the data.")
```

```python
# How many cards do we have recorded in our dataset?

# Using sum() on the Cards column
total_cards = df["Cards"].sum()

print(f"There are {total_cards} total cards registered.")

```

```python
# How many females do we have information for vs how many males?

# Creating 2 refined frames based on gender, and using len() to find
# how many samples there are in each
males = len(df[df["Gender"] == "Male"])
females = len(df[df["Gender"]== "Female"])

# f-string to print
print(f"There are {females} Females and {males} Males in our data")
```

```python
# Once again, this can be acheived with the following line;

print(df.groupby("Gender")["Gender"].count())
```
